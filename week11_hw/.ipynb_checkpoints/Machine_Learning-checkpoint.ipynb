1. K-Means Clustering on Protein Data & Elbow Plot

```python
# set your working director to analysis data
import os
os.chdir('/Users/colinyeo/desktop/School/qbio_490_colin/analysis_data/')

# import cptac, numpy, and pandas
import cptac
import numpy as np
import pandas as pd

# download the Brca dataset
cptac.download(dataset="BRCA")
brca = cptac.Brca()

# extract the data
clinical_data = brca.get_clinical()
transcriptomic_data = brca.get_transcriptomics()
protein_data = brca.get_proteomics()
protein_data.columns = protein_data.columns.get_level_values(0) 
```

```python
# getting some functions
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# drop all genes with NAs
protein_data = protein_data.dropna(axis="columns")

# let's choose A1BG expression for x-axis and A2M expression for y-axis
d = {'A1BG':protein_data.iloc[:, 0],'A2M':protein_data.iloc[:, 1]}
df = pd.DataFrame(d, columns=['A1BG','A2M'])

# elbow plot - 3 clusters are optimal
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)

# commented this out because it was clashing with the actual k-means scatter plot
plt.plot(range(1, 11), wcss)
plt.show()

# clear
plt.clf()

# do the k-means 
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0).fit(df)
sns.scatterplot(data = df, x = "A1BG", y = "A2M", hue = kmeans.labels_).set(title = "Regulation of A2M vs. A1BG Proteins")
plt.show()
```

2. PCA on RNA data 

```python
# get functions
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# get rna clinical data from csv & make a new df with just barcode + stage
rna_clinical = pd.read_csv("brca_rna_clinical_data.csv")
clinical_dat = {'Barcode':rna_clinical.loc[:, "barcode"],'Stage':rna_clinical.loc[:, "paper_pathologic_stage"]}
target_df = pd.DataFrame(clinical_dat, columns=['Barcode','Stage'])
barcode_dict = pd.Series(target_df.Stage.values,index=target_df.Barcode).to_dict()

# get rna counts data from csv, perform dimensionality reduction first
rna_counts = pd.read_csv("brca_rna_count_data.csv")
rna_counts = rna_counts.iloc[:, 2:len(rna_counts.columns)]

# transpose rna_counts for convenience
rna_counts = rna_counts.T
rna_counts['barcodes'] = rna_counts.index
rna_counts['stage'] = rna_counts['barcodes'].map(barcode_dict)
rna_counts = rna_counts.drop(['barcodes'],axis =1)

# last, drop any rows with NA in them
rna_counts = rna_counts.dropna(axis="rows")

# at this point, we have a properly formatted rna_counts df to undergo pca

# standardize the rna counts
x = rna_counts.loc[:, rna_counts.columns != 'stage'].values
x = StandardScaler().fit_transform(x)

# do PCA now
pca = PCA(n_components = 2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents, columns = ['pc1', 'pc2'])

# format for plotting
principalDf['stage'] = rna_counts['stage'].values

# last, plot it
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('pc1', fontsize = 15)
ax.set_ylabel('pc2', fontsize = 15)
ax.set_title('2 component PCA', fontsize = 20)
targets = pd.unique(principalDf.loc[:, "stage"])
colors = ['r', 'g', 'b', 'y']
for target, color in zip(targets,colors):
    indicesToKeep = principalDf['stage'] == target
    ax.scatter(principalDf.loc[indicesToKeep, 'pc1']
               , principalDf.loc[indicesToKeep, 'pc2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
